```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fifa = pd.read_csv('working_data/fifa22p.csv', low_memory=False)
```

```python
pd.Series(fifa.columns).head(60)
```

```python
top_wages = fifa.groupby('nationality_name')['wage_eur'].max()
wmtop = top_wages[top_wages > 100000]
plt.figure(figsize = (15, 5))
plt.bar(wmtop.index, wmtop, color = '#FF3660')
plt.xticks(rotation = '90')
plt.title('Top Paid Footballers by Nationality (2021-2022). Can You Guess Them All?')
plt.show()
```

```python

en = fifa[fifa['nationality_name'] == 'England'].head(850)
fr = fifa[fifa['nationality_name'] == 'France'].head(850)
de = fifa[fifa['nationality_name'] == 'Germany'].head(850)
es = fifa[fifa['nationality_name'] == 'Spain'].head(850)
iy = fifa[fifa['nationality_name'] == 'Italy'].head(850)

en.index = np.arange(0, 850, dtype = 'int')
fr.index = np.arange(0, 850, dtype = 'int')
de.index = np.arange(0, 850, dtype = 'int')
es.index = np.arange(0, 850, dtype = 'int')

weur = pd.DataFrame()

weur['England'] = en['wage_eur']
weur['France'] = fr['wage_eur']
weur['Germany'] = de['wage_eur']
weur['Spain'] = es['wage_eur']
weur['Italy'] = iy['wage_eur']


plt.figure(figsize = (15, 5))
sns.boxplot(data = weur, palette = 'afmhot')
plt.plot()
```

```python
plt.figure(figsize = (15, 5))
sns.set_context('notebook')
sns.violinplot(data = weur, palette = 'cool', bw = 0.2)
plt.ylabel('Wages in Euros')
plt.plot()
```

```python
en = fifa[(fifa['nationality_name'] == 'England') & (fifa['age'] <= 21)].head(200)
fr = fifa[(fifa['nationality_name'] == 'France') & (fifa['age'] <= 21)].head(200)
de = fifa[(fifa['nationality_name'] == 'Germany') & (fifa['age'] <= 21)].head(200)
es = fifa[(fifa['nationality_name'] == 'Spain') & (fifa['age'] <= 21)].head(200)

en.index = np.arange(0, 200, dtype = 'int')
fr.index = np.arange(0, 200, dtype = 'int')
de.index = np.arange(0, 200, dtype = 'int')
es.index = np.arange(0, 200, dtype = 'int')


pot_eur = pd.DataFrame()

pot_eur['England'] = en['potential']
pot_eur['France'] = fr['potential']
pot_eur['Germany'] = de['potential']
pot_eur['Spain'] = es['potential']

pot_eur
```

```python
plt.figure(figsize = (15, 5))
sns.set_context('notebook')
sns.swarmplot(data = pot_eur, palette = 'coolwarm')
plt.ylabel('Club Jersey Numbers')
plt.title('Best Under 21 Young Talents in Europe')
plt.plot()
```

```python
plt.figure(figsize = (15, 5))
sns.set_context('paper')
sns.violinplot(data = pot_eur, palette = 'spring', bw = 0.2)
sns.swarmplot(data = pot_eur, color = 'black')

plt.ylabel('Potential Rating')
plt.title('Categorical Dist. of Top U21 Talents in Europe')
plt.show()
```

```python
plt.figure(figsize = (19, 15))
sns.set_context('poster')
sns.stripplot(data = jnum_eur, palette = 'bone', jitter = 0.35, alpha = 0.65, size = 8)
plt.ylabel('Club Jersey Numbers')
plt.title('Most Popular Jersey Numbers by Countries. Top 850 Players')
plt.plot()
```
